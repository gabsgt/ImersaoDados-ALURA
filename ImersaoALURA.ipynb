```python
import numpy as np

corr = dados.loc[:, 'g-0': 'g-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

#dado um x, eu consigo predizer um y: f(x) = y - sendo x as variáveis
#selecionar apenas as assinaturas celulares = genes (g-) + os tipos celulares (c-)

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
#antes de treinar todo o dataset que temos disponivel, precisamos separar ele entre dataset de treino, e dataset de teste. 
#Pq se treinarmos todo o dataset, não vamos saber se o algoritmo está acertando, ou não.
#proporção de quanto deve ser separado depende do volume de dados disponível: alguns padrões - 20% para teste (dtset pequeno)
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=354)

modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino , y_treino) #treinamento do modelo
modelo_rlogistica.score(x_teste, y_teste) #calcular a acurácia do modelo
#o output é a proporção de quanto o modelo acertou a predição
#como melhorar essa acurácia? podemos mudar os parametros
```

```python
#para saber se a acurácia é suficiente, precisamos de algum parametro de comparação
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier('most_frequent') #definir a estratégia
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=354)

modelo_arvore = DecisionTreeClassifier(max_depth = 3)
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)
```

```python
from sklearn import tree

fig, ax = plt.subplots(figsize=(15, 10), facecolor='k')
tree.plot_tree(modelo_arvore,
               ax=ax,
               fontsize=10,
               rounded=True,
               filled=True,
               feature_names=x_treino.columns,
               class_names=['Não Ativado', 'Ativado'])

plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=354)

teste=[]
treino=[]
for i in range(1,15):
    modelo_arvore = DecisionTreeClassifier(max_depth = i)
    modelo_arvore.fit(x_treino, y_treino)
    teste.append(modelo_arvore.score(x_teste, y_teste))
    treino.append(modelo_arvore.score(x_treino, y_treino))
```

```python
teste
```

```python
treino
```

```python
sns.lineplot(x=range(1,15), y=teste, label='teste')
sns.lineplot(x=range(1,15), y=treino, label='treino')

#o resultado aqui foi de overfit = quando o modelo fica tão bom em adivinhar os dados de treino, e ele não consegue testar os dados de teste
```
